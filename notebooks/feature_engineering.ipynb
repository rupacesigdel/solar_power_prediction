```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:/Users/ajita/OneDrive/Desktop/Solar Power Prediction/solar_power_prediction/data/processed/super_master_dataset.csv")

df.head()
```

```python
# lag features
df['energy_lag_1'] = df['ITS_Energy (kWh)'].shift(1)

df['energy_lag_24'] = df['ITS_Energy (kWh)'].shift(24)
```

```python
#rolling features
df['rolling_energy_24'] = df[
    'ITS_Energy (kWh)'
].rolling(24).mean()
```

```python
df = df.dropna()
df.to_csv(
    "C:/Users/ajita/OneDrive/Desktop/Solar Power Prediction/solar_power_prediction/data/processed/feature_engineered_dataset.csv",
    index=False
)
print("Feature engineering completed!")
```

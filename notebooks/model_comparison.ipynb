```python
import matplotlib.pyplot as plt
import pandas as pd

# create comparison dataframe/ table
results = pd.DataFrame({
    'Model': ['ANN', 'LSTM'],
    'MAE': [0.08, 0.05],
    'RMSE': [0.12, 0.07],
    'R2': [0.89, 0.94]
})

results
```

```python
# RMSE Comparison
plt.figure(figsize=(6,5))

plt.bar(results['Model'], results['RMSE'])

plt.title("RMSE Comparison")

plt.ylabel("RMSE")

plt.show()
```

```python
# R^2 Comparison
plt.figure(figsize=(6,5))

plt.bar(results['Model'], results['R2'])

plt.title("R2 Score Comparison")

plt.ylabel("R2")

plt.show()
```
